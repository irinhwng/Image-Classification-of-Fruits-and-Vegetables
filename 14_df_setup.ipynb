```python
import pandas as pd
```

```python
#reading in csv files
apple = pd.read_csv('./datasets/dataset_with_kaggle_pics/apple.csv')
lemon = pd.read_csv('./datasets/dataset_with_kaggle_pics/lemon.csv')
bellpepper = pd.read_csv('./datasets/dataset_with_kaggle_pics/bellpepper.csv')
onion = pd.read_csv('./datasets/dataset_with_kaggle_pics/onion.csv')
pear = pd.read_csv('./datasets/dataset_with_kaggle_pics/pear.csv')
strawberry = pd.read_csv('./datasets/dataset_with_kaggle_pics/strawberry.csv')
tomato = pd.read_csv('./datasets/dataset_with_kaggle_pics/tomato.csv')
```

```python
#sanity check
print (f'apple null count {apple.isnull().sum().sum()}')
print (f'lemon null count {lemon.isnull().sum().sum()}')
print (f'bellpepper null count {bellpepper.isnull().sum().sum()}')
print (f'onion null count {onion.isnull().sum().sum()}')
print (f'pear null count {pear.isnull().sum().sum()}')
print (f'strawberry null count {strawberry.isnull().sum().sum()}')
print (f'tomato null count {tomato.isnull().sum().sum()}')
```

```python
#combine all the dataframes together
frames = [apple, lemon, bellpepper, onion, pear, strawberry, tomato]

df = pd.concat(frames)
df.head()
```

```python
#extra sanity check
df.isnull().sum().sum()
```
