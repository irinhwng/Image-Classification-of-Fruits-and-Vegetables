```python
import cv2
import numpy as np
import pandas as pd
```

```python
#setting up columns for df of pixels
pixel_list = []
for i in range(30_000):
    pixel_name = f'pixel{i}'
    pixel_list.append(pixel_name)
```

```python
#setting up columns names to be pixel list (from pixel0 to pixel29999)
tomato = pd.DataFrame(columns = pixel_list )
```

```python
#sanity check
tomato
```

```python
produce_amount = 8_000
for i in range(1, produce_amount+ 1):
    try:
        # set up image path 
        # remember, we're getting our images from 03_{produce_name}_resize from the images folder 
        img_path = f'./images/03_tomato_resize/{i}_tomato.jpg'
        img = cv2.imread(img_path, 1) # read image as color. Set second parameter to 1 if rgb is required 
        
        # this code turns the image to 30,000 numbers since 100*100*3 = 30,000
        row = img.ravel()
        
        #this is a sanity check
        # all images are 100x100x3 so this code would run for every image
        if len(row) == 30_000:
            
            #this code appends the new row of pixels to the dataframe 
            tomato.loc[i] = row
            print(f'added image {i}')
    except: 
        pass
```
